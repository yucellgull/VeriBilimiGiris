```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yaslistesi = [10,20,25,32,34,37,40,46,49,51]
```

```python
kilolistesi = [15,45,60,65,70,77,88,90,92,100]
```

```python
numpyyaslistesi = np.array(yaslistesi)
numpykilolistesi = np.array(kilolistesi)
```

```python
numpyyaslistesi
```

```python
numpykilolistesi
```

```python
plt.plot(numpyyaslistesi,numpykilolistesi,"b")
plt.xlabel("Yaş")
plt.ylabel("Kilo")
plt.title("Kilo'nun yaşa göre grafiği")
#plt.show()
```

```python
numpydizi1 = np.linspace(0,10,20)
```

```python
numpydizi1
```

```python
numpydizi2 = numpydizi1**3
```

```python
numpydizi2
```

```python
plt.plot(numpydizi1,numpydizi2,"b*-")
```

```python
plt.subplot(1,2,1)
plt.plot(numpydizi1,numpydizi2,"r*-")
plt.subplot(1,2,2)
plt.plot(numpydizi1,numpydizi2,"g*-")
```

```python
bnmfigure = plt.figure()
figureaxes = bnmfigure.add_axes([0.2,0.2,0.9,0.9])
figureaxes.plot(numpydizi1,numpydizi2,"g")
figureaxes.set_xlabel("x ekseni")
figureaxes.set_ylabel("y ekseni")
figureaxes.set_title("başlık")
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.3,0.3,0.3,0.3])

eksen1.plot(numpydizi1,numpydizi2,"g")
eksen1.set_xlabel("x ekseni")
eksen1.set_ylabel("y ekseni")
eksen1.set_title("ana başlık")

eksen2.plot(numpydizi1,numpydizi2,"r")
eksen2.set_xlabel("x ekseni")
eksen2.set_ylabel("y ekseni")
eksen2.set_title("kücük başlık")
```
